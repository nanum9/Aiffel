##  프로젝트 (1) load_digits : 손글씨를 분류

```python
# 우선 라이브러리 버전을 확인
import sklearn

print(sklearn.__version__)
```

```python
# 손글씨 이미지를 제대로 0~9까지 열가지 카테고리로 분류해 본다
# 필요한 모듈 import
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 데이터 준비, load_digits 메서드를 사용
digit = load_digits()

print(dir(digit))
```

```python
# digit 데이터셋의 데이터 파악
digit.keys()
```

```python
# 가장 중요한 데이터는 digit_data 변수에 저장
digit_data = digit.data

print(digit_data.shape) # 1797개의 데이터가 각각 64개의 정보를 담고 있음
```

```python
digit_data[0]
#샘플로 하나의 데이터만 확인.1797개의 데이터 중 첫 번째 데이터 출력. 
```

```python
digit_label = digit.target #타겟 정보 . keys에서 확인한 정보 중 target을 따로 digit_label 변수에 저장
print(digit_label.shape)
digit_label
```

```python
# 각 feature에 대한 설명
digit.feature_names
```

```python
import pandas as pd

print(pd.__version__) # Q 이거 굳이 해야되나?
```

```python
# digit.feature_names을 컬럼명(세로열 목록)으로 하는 DataFrame 자료형으로 변환해서 digit_df 변수에 저장
import pandas as pd

digit_df = pd.DataFrame(data=digit_data, columns=digit.feature_names)

digit_df
```

```python
# digit_df에 label이라는 컬럼을 새로 추가
digit_df["label"] = digit.target

digit_df
```

```python
#데이터 Describe 

digit_df.describe()
## 데이터 Describe한 data봐도 잘 이해 안감.
```

```python
# 머신러닝 모델을 학습시키려면  training dataset과 test dataset나누는 작업이 필요.
from sklearn.model_selection import train_test_split 


X_train, X_test, y_train, y_test = train_test_split(digit_data, # X는 data, y는 정답, 쪼개 X_train, X_test, y_train, y_test에 저장;                                              
                                                    digit_label, 
                                                    test_size=0.2, 
                                                    random_state=15)       


print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))  # len은 배열의 길이를 출력    
```

## 의사결정나무, Random Forest , SVM , SGD Classifier, Logistic Regression 통해 학습시키기

```python
## 1. 의사결정나무로 학습시킨 결과 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

decision_tree = DecisionTreeClassifier(random_state=15) 
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test) 

print(classification_report(y_test, y_pred))
```

```python
# Random Forest로  학습시킨 결과
from sklearn.ensemble import RandomForestClassifier 

random_forest = RandomForestClassifier(random_state=15) 
random_forest.fit(X_train, y_train) 
y_pred = random_forest.predict(X_test) 

print(classification_report(y_test, y_pred)) 
```

```python
# SVM 분류기로  학습시킨 결과
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train) 
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred)) 
```

```python
# SGD 분류기로  학습시킨 결과
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier() 
sgd_model.fit(X_train, y_train) 
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Logistic 회귀로  학습시킨 결과 
from sklearn.linear_model import LogisticRegression 

logistic_model = LogisticRegression(max_iter=5000)
logistic_model.fit(X_train, y_train) 
y_pred = logistic_model.predict(X_test) 

print(classification_report(y_test, y_pred))
```

## 평가하기 
정밀도(precision) : 양성이라고 예측한 값 중 실제 양성인 값의 비율, 5가지 학습결과를 살펴보니  precision이 비교적 높게 나왔음.

##  프로젝트 (2) load_wine : 와인을 분류

```python
# 필요한 모듈 import
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 데이터 준비, load_wine 메서드를 사용
from sklearn.datasets import load_wine

wine = load_wine()

print(dir(wine))
```

```python
# wine 데이터셋의 데이터 파악
wine.keys()
```

```python
# 가장 중요한 데이터는 wine_data 변수에 저장
wine_data = wine.data

print(wine_data.shape) # 178개의 데이터가 각각 13개의 정보를 담고 있음
```

```python
# wine_data 중 샘플로 하나의 데이터만 확인.178개의 데이터 중 첫 번째 데이터 출력. 
wine_data[0]
```

```python
#타겟 정보 . keys에서 확인한 정보 중 target을 따로 wine_label 변수에 저장
wine_label = wine.target

print(wine_label.shape)
wine_label
```

```python
# 각 feature에 대한 설명 
wine.feature_names 
```

```python
# wine.feature_names을 컬럼명(세로열 목록)으로 하는 DataFrame 자료형으로 변환해서 wine_df 변수에 저장
wine_df = pd.DataFrame(data=wine_data, columns=wine.feature_names)

wine_df
```

```python
# wine_df에 label이라는 컬럼을 새로 추가
wine_df["label"] = wine.target

wine_df
```

```python
#데이터 Describe : 평균, 표준편차등 통계 데이터 보여줌.
wine_df.describe()
```

```python
# training dataset과 test dataset를 분리하는 작업
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=15)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

## 1.의사결정나무, 2. Random Forest, 3. SVM, S4. GD Classifier, 5. Logistic Regression 통해 학습시키기

```python
## 1. 의사결정나무로 학습시킨 결과 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

decision_tree = DecisionTreeClassifier(random_state=15) 
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test) 

print(classification_report(y_test, y_pred))
```

```python
# Random Forest로  학습시킨 결과
from sklearn.ensemble import RandomForestClassifier 

random_forest = RandomForestClassifier(random_state=15) 
random_forest.fit(X_train, y_train) 
y_pred = random_forest.predict(X_test) 

print(classification_report(y_test, y_pred)) 
```

```python
# SVM 분류기로  학습시킨 결과
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train) 
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred)) 
```

```python
# SGD 분류기로  학습시킨 결과
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier() 
sgd_model.fit(X_train, y_train) 
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Logistic 회귀로  학습시킨 결과 
from sklearn.linear_model import LogisticRegression 

logistic_model = LogisticRegression(max_iter=5000)
logistic_model.fit(X_train, y_train) 
y_pred = logistic_model.predict(X_test) 

print(classification_report(y_test, y_pred))
```

## 평가하기 
recall : 실제 True인 것 중에서 모델이 True라고 예측한 것의 비율, recall 수치가 5가지 학습중 잘 나왔으며 와인성분, 점도등의 data를 주었을때 recall이 가장 정확도가 올라감.

## 프로젝트 (3) load_breast_cancer : 유방암 여부 진단

```python
# 필요한 모듈 import
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# 데이터 준비, load_breast_cancer 메서드를 사용
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()

print(dir(breast_cancer))
```

```python
# breast_cancer 데이터셋의 데이터 파악
breast_cancer.keys()
```

```python
# 가장 중요한 데이터는 breast_cancer _data 변수에 저장
breast_cancer_data = breast_cancer.data

print(breast_cancer_data.shape) # 569개의 데이터가 각각 30개의 정보를 담고 있음
```

```python
# breast_cancer_data 중 샘플로 하나의 데이터만 확인.569개의 데이터 중 첫 번째 데이터 출력. 
breast_cancer_data[0]
```

```python
#타겟 정보 . keys에서 확인한 정보 중 target을 따로 breast_cancer_label 변수에 저장
breast_cancer_label = breast_cancer.target

print(breast_cancer_label.shape)
breast_cancer_label
```

```python
# feature에 대한 설명 
breast_cancer.feature_names 
```

```python
# breast_cancer.feature_names을 컬럼명(세로열 목록)으로 하는 DataFrame 자료형으로 변환해서 breast_cancer_df 변수에 저장
breast_cancer_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer.feature_names)

breast_cancer_df
```

```python
# breast_cancer_df에 label이라는 컬럼을 새로 추가
breast_cancer_df["label"] = breast_cancer.target

breast_cancer_df
```

```python
# breast_cancer_df의 평균, 표준편차등 통계 data를 보여준다. 
breast_cancer_df.describe()
```

```python
# training dataset과 test dataset를 분리하는 작업
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=15)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

```python
## 1.의사결정나무, 2. Random Forest, 3. SVM, S4. GD Classifier, 5. Logistic Regression 통해 학습시키기
```

```python
## 1. 의사결정나무로 학습시킨 결과 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

decision_tree = DecisionTreeClassifier(random_state=15) 
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test) 

print(classification_report(y_test, y_pred))
```

```python
# Random Forest로  학습시킨 결과
from sklearn.ensemble import RandomForestClassifier 

random_forest = RandomForestClassifier(random_state=15) 
random_forest.fit(X_train, y_train) 
y_pred = random_forest.predict(X_test) 

print(classification_report(y_test, y_pred)) 
```

```python
# SVM 분류기로  학습시킨 결과
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train) 
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred)) 
```

```python
# SGD 분류기로  학습시킨 결과
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier() 
sgd_model.fit(X_train, y_train) 
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Logistic 회귀로  학습시킨 결과 
from sklearn.linear_model import LogisticRegression 

logistic_model = LogisticRegression(max_iter=5000)
logistic_model.fit(X_train, y_train) 
y_pred = logistic_model.predict(X_test) 

print(classification_report(y_test, y_pred))
```
